```python
region = {}
subregion = {}

fails = []

regions = sorted(deaths['Country/Region'].unique())
for r in regions:
  try:
    country = CountryInfo(r)
    region[r] = country.region()
    subregion[r] = country.subregion()
  except:
    fails.append(r)
    print('No data for', r)
```

```python
region['Cabo Verde'] = 'Africa'
region['Comoros (the)'] = 'Africa'
region['Congo (Brazzaville)'] = 'Africa'
region['Congo (Kinshasa)'] = 'Africa'
region["Cote d'Ivoire"] = 'Africa'
region['Eswatini'] = 'Africa'
region['Gambia'] = 'Africa'
region['Sao Tome and Principe'] = 'Africa'



subregion['Cabo Verde'] = 'Western Africa'
subregion['Comoros (the)'] = 'Southern Africa'
subregion['Congo (Brazzaville)'] = 'Middle Africa'
subregion['Congo (Kinshasa)'] = 'Middle Africa'
subregion["Cote d'Ivoire"] = 'Western Africa'
subregion['Eswatini'] = 'Southern Africa'
subregion['Gambia'] = 'Western Africa'
subregion['Sao Tome and Principe'] = 'Middle Africa'
```

```python
def addColumns(df, region=region, subregion=subregion):
    # Adding in population as a column
    df['Region'] = df['Country/Region'].map(region)
    df['Subregion'] = df['Country/Region'].map(subregion)
    return df
```

```python
cases = addColumns(cases)
recover = addColumns(recover)
deaths = addColumns(deaths)
```

```python
# Re-arrange the data
cols = list(cases.columns)
cases = cases[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
recover = recover[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
deaths = deaths[cols[:4]+['Region','Subregion']+cols[4:-2]] # Better ordering
```

```python
display(cases.head(), recover.head(), deaths.head())
```

```python
cases_africa = cases[cases['Region']=='Africa']
recover_africa = recover[recover['Region']=='Africa']
deaths_africa = deaths[deaths['Region']=='Africa']
```

```python
display(cases_africa.head(), recover_africa.head(), deaths_africa.head())
```

```python
cases_africa.to_csv('data/africa_cases.csv', index=False)
recover_africa.to_csv('data/africa_recover.csv', index=False)
deaths_africa.to_csv('data/africa_deaths.csv', index=False)
```
